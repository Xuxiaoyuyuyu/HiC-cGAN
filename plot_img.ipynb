```python
import os
import numpy as np 
from matplotlib import pyplot as plt
import cv2
def rgb2gray(rgb):
    return np.dot(rgb[..., :3], [0, 0, 1])
```

```python
num = '002' # 001~655
            # 效果较好的有 002 231
cmp = 'Reds'
hr=rgb2gray(cv2.imread('./test/'+ num +'_hr.jpg'))
sr=rgb2gray(cv2.imread('./test/'+ num +'_sr.jpg'))
lr=rgb2gray(cv2.imread('./test/'+ num +'_lr.jpg'))

plt.figure(figsize=(100,100))

plt.subplot(1,3,1)
plt.imshow(hr, cmap=cmp , vmax=255,vmin=0)
plt.subplot(1,3,2)

plt.imshow(sr, cmap=cmp , vmax=255,vmin=0)
plt.subplot(1,3,3)
plt.imshow(lr, cmap=cmp , vmax=255,vmin=0)
# plt.savefig("./fig.png")
plt.show()
```
